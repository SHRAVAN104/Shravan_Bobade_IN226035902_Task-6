## **Problem Statement 1:Smart Parking Lot Management System**

```python
def smart_parking_system(capacity, logs):
    parked = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        peak = max(peak, parked)
    
    print(f"Currently Parked Vehicles: {parked}")
    
    if parked > capacity:
        print("Parking Status: Full (Over Capacity Alert!)")
    elif parked == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
    
    print(f"Peak Usage: {peak}")

# Sample Test
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

## **Problem Statement 2:Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km
    
    # Traffic delay
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8
    
    # Weather delay
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20
    
    print(f"Estimated Delivery Time: {base_time} minutes")

# Sample Test
estimate_delivery_time(8, "High", "Rainy")
```

## **Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    
    print(f"Occupancy: {occupancy:.0f}%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

# Sample Test
seat_occupancy(200, [1]*150)
```

## **Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    
    print(f"Average CPU Load: {avg_cpu:.0f}%")
    
    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Sample Test
classify_server_load([45, 60, 70, 85, 90])
```

## **Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def classroom_monitor(resources):
    overused = []
    
    for resource, hours in resources.items():
        if hours > 8:   # Threshold for overuse
            overused.append(resource)
    
    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")

# Sample Test
classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

## **Problem Statement 6: Online Event Registration Capacity Controller**

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Sample Test
event_registration(100, 105)
```
